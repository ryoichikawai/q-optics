# Quantum measurement of photons

Quantum measurement is totally different from classical measurement.  Quantum measurement processes cannot be described by Schr&ouml;dinger equation. We need a separate theory for quantum measurement, which has not been fully understood.  For matter  particles such as electrons, physical quantities are defined as operators (observables) in a Hilbert space and the measurement process is projective (projection-values measurement or PVM).  After the measurement, the state of the particles jumps to an eigenstate of the observable.  If the same measurement is repeated immediately after the first one, the same result is obtained.  Measurement of photons is quite different. Since photons annihilate upon the measurement, it is a destructive measurement and cannot be described by PVM.  The repeated measurement is not possible.  Mathematically, it can be described by positive operator-valued measurement (POVM).  However, the detection mechanism of photons are still poorly understood.  Unfortunately, there is not much progress since the pioneering theory constructed by R. J. Glauber in 1960s.{cite:p}`Glauber1969`
