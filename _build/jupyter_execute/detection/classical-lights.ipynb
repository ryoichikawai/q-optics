# Detection of classical lights

## Intensity of classical lights

In the classical theory, light is electromagnetic wave which is continuous in space and time. Mathematically, it is expressed as vector field $\mathbf{E}(\mathbf{x},t)$ and $\mathbf{B}(\mathbf{x},t)$ Its character is measured with two temporal correlation functions, the first-order and second-order correlation defined respectively by

$$
g^{(1)}(\tau) = \frac{\langle\mathbf{E}^{*}(t) \cdot \mathbf{E}(t+\tau)\rangle_t}{\langle\mathbf{E}^{*}(t) \cdot \mathbf{E}(t)\rangle_t} = \frac{\langle\mathbf{E}^{*}(t) \cdot \mathbf{E}(t+\tau)\rangle_t}{\langle I(t) \rangle_t}
$$(eq_g1-classical)

and

$$
g^{(2)}(\tau) = \frac{\langle\left[\mathbf{E}^{*}(t) \cdot \mathbf{E}(t)\right]\left[\mathbf{E}^{*}(t+\tau) \cdot \mathbf{E}(t+\tau)\right]\rangle_t}{(\langle\mathbf{E}^{*}(t) \cdot \mathbf{E}(t)\rangle_t)^2}
= \frac{\langle I(t) I(t+\tau)\rangle_t}{(\langle I(t) \rangle_t)^2}
$$(eq_g2-classical)

where $I(t)=\mathbf{E}^{*}(t)\cdot \mathbf{E}(t)$ is the intensity of the field and $\langle\cdots\rangle_t$ indicates average over time $t$.   We have use the complex expression of the electric fields for mathematical convenience.

Usually, the correlation vanishes for a large $\tau$ and $\langle\mathbf{E}^{*}(t) \cdot \mathbf{E}(t+\tau)\rangle_t \rightarrow (\langle \mathbf{E}(t) \rangle_t)^2 = 0$ and $\langle I(t) I(t+\tau)\rangle_t = (\langle I(t) \rangle_t)^2$, from which we obatin.  

$$
\lim_{\tau\rightarrow\infty} g^{(1)}(\tau) = 0
$$

and

$$
\lim_{\tau\rightarrow\infty} g^{(2)}(\tau) = 1
$$

:::{note}
It is very important to keep it in our mind that the measurement is done at a certain fixed point in space, which is not explicitly written in the above expression. $g^{(1)}(\tau)$ and $g^{(2)}(\tau)$ can 
strongly depend on the location.
:::

The fist-order correlation provides information about the oscillatory behavior of the field through its Fourier transform known as power spectrum (Wiener–Khinchin theorem):

$$
I(\omega) = \frac{1}{2\pi} \int_{-\infty}^\infty g^{(1)}(\tau) e^{-i \omega \tau} d\tau
$$

In other words, the first-order correlation gives us frequency distribution of the light.

## No-delay correlation

No-delay ($\tau=0$) second-order correlation $g^(2)(0)$ becomes a very important quantity to investigate both experimentally and theoretically. 
While $g^{(1)}(0)=1$ is always true, the value of $g^{(2)}(0)$ varies depending on the character of light.  Introducing mean intensity $I = \langle I(t) \rangle_t$, we express the fluctuation of intensity by a variance:  

$$
(\Delta I)^{2} = \langle I^{2}(t) \rangle_t - I^{2}
$$(eq:

we obtain

$$
g^{(2)}(0) = \frac{\langle I^{2}(t)\rangle_t}{(\langle I(t) \rangle_t)^2} = 1 + \left(\frac{\Delta I}{I}\right)^{2}
$$

which suggests that $g^{(2)}(0) \ge 1$. 

   

The positive correlation suggests that  This was first confirmed by Hanbury Brown and Twiss.{cite:p}`HanburyBrown1956`  with the advent of modern experimental technique,  cleaner results have been obtained.{cite:p}`Tan2014,Tan2017`    If and only if $\Delta I =0$, then $g^{(2)}(0)=1$.  This means that if an experiment data shows $g^{(2)}(0)=1$ then the light intensity is not fluctuating.
It turns out that negative value of $g^{(2)}(0)$ has been observed{cite:p}`Kimble1977,Grangier1986` in violation of the above result.  Such results can be explained only by quantum optics.


## Hanbury Brown and Twiss effect - the emergence of quantum optics

In 1954, Hanbury Brown and Twiss (HBT) proposed a new type of interferometer.{cite:p}`HanburyBrown1954` A light beam is split in half, a half of the beam ($L_1$) goes to a detector $D_1$ and the other half ($L_2$) to another detector $D_2$.  The intensity of light usually fluctuates in time.  At detector $D_i$, the flustuation is defined as $\Delta I_i (t) = I_i(t) - \langle I \rangle_t$ where $\langle \cdots \rangle_t$ indicate temporal average over a period longer than the time scale of the fluctuation.  Hence mean intensity does not depend on time.  Their interferometer measures the temporal correlation of intensity fluctuation between two detectors.   

$$
\langle \Delta I_{1}(t) \Delta I_{2}(t+\tau)\rangle_{t} = \langle I_{1}(t) I_{2}(t+\tau)\rangle_{t}
-\langle I \rangle^{2}
$$(eq_HBT)

where we assume that $\langle I_{1}(t) \rangle_t = \langle I_{2}(t) \rangle_t \equiv \langle I \rangle$. 

Equation {eq}`eq_HBT` looks similar but slightly different because the two intensities are measured by different detectors.  It is actually measuring $g^{(2)}(t)$ if the dead period of detectors is taken into account. Even the modern detectors can't record two photons arriving nearly at the same time.  Hence, $\langle I_i(t) I_i(t) \rangle$ vanishes when two photons arrive at the same time. Although the detector can record $\langle I_i(t) I_i(t+\tau) \rangle$ if $\tau$ is larger than the dead time, we need to discard it cancel the counting loss during the dead time.  We lose 50\% of light but the remaining part measures $g^{(2)}(t)$.
In 1956, they carried out the proposed experiment and observed strong correlation between the two detectors, especially at no time-delay.{cite:p}`HanburyBrown1956`   The result is known as Hanbury Brown and Twiss (HBT) effect.  Essentially, this experiment confirmed that $g^(2)(0) > 1$.

The results were seemingly obvious from the classical theory since the same light just took two separate paths. Strangely, physicists at that time struggled to explain the results using quantum mechanics.
Many of them learned quantum mechanics from  Dirac's book "The principle of Quantum Mechanics" which said  "Each photon interferes only with itself."  Based on this, they wonder how two photons can arrive at two different places at the same time beyond the accidental coincidence?  The two detectors are measuring two different group of photons which are supposedly independent to each other. 

The controversy was resolved by Fano in 1961.{cite:p}`Fano1961`  Two photons can be coherent if they are indistinguishable.  Suppose that photons $\psi_{1}$ and $\psi_{2}$ are detected respectively by detectors $D_1$ and $D_2$.  Let us write the detection probability symbolically $|\langle D_{2} D_{2}| \psi_{1} \psi_{2}\rangle|^2$. There is another possibility that photon $\psi_{1}$ goes to $D_2$ and $\psi_2$ to $D_{1}$. In other words, the state of photons should be $\frac{1}{\sqrt{2}} \left(\psi_{1}\psi_{2} + \psi_{2}\psi_{1}\right)$ since they are bosons.  Then, the probability that the two detectors find photons are

$$
\frac{1}{2} \left | \langle D_{1} D_{2} | (|\psi_{1}\psi_{2} + \psi_{2}\psi_{1}\rangle ) \right |^2 = \frac{1}{2} \left(|\langle D_{1} D_{2}|\psi_{1}\psi_{2}\rangle|^2 + |\langle D_{1} D_{2}|\psi_{2}\psi_{1}\rangle\rangle|^2 \right) + \text{Re} \langle D_{1} D_{2} | \psi_{2}\psi_{1} \rangle \langle \psi_{1}\psi_{2} | D_{1} D_{2}\rangle
$$

Due to the interference term, the probability increases.   Two photons can jointly interfere with themselves without violating the Dirac's statement. Note that if the particles were fermions, then the interference term would be negative and thus the correlation is reduced, which has been confirmed.{cite:p}`Henny1999,Jeltes2007`

Immediately after the discovery of the HBT effect, physicists realized that a better theory on the coherence of light was urgently needed and modern quantum optics was born.{cite:p}`Glauber2007`. While the HBT effect can be easily explained by the classical theory of light, it initiated the modern quantum optics and even now their method is widely used.
