# Quantum coherence functions

In {numref}`sec:FCS`, we learn three different types of photon counting statistics.  The key point was that for the waiting time distribution, the magnitude of fluctuation with respect to the mean determines the character of the light. We also see in {numref}`sec:classical-lights` that $g^{(2)}(0)$ dictates the variance of intensity at classical lights.  Hence, the correlation function $g^{(2)}(\tau)$ for quantum lights should be a good indicator of the photon counting statistics.  In this chapter, we derive a general mathematical expression of the first/second-order coherence functions and discuss their general properties.  The case studies with simple models follows in separate sections.


## Relation with photons and electric fields.

Suppose that the experimental setup can see only a single mode. 
Now, the electric field corresponding to the mode is expressed by

$$
\begin{eqnarray}
E^{(+)}(t) &=& \mathcal{E} a(t)= \mathcal{E}_0 a e^{-i \omega t} \\
E^{(-)}(t) &=& \mathcal{E} a^{\dagger}(t) = \mathcal{E}_0 a^{\dagger} e^{i \omega t}
\end{eqnarray}
$$

where $a(t)$ and $a^{\dagger}(t)$ are annihilation and creation operators in the Heisenberg picture and $a$ and $a^{\dagger}$ are corresponding operators in the Schr&ouml;dinger picture.     $\mathcal{E}$ is a proportional constant.  Since the photons are in a single mode, the direction of polarization is omitted.  Then, the intensity of the field is given by

$$
I(t) = \langle E^{(-)}(t) E^{(+)}(t)\rangle = \mathcal{E}^{2} \langle a^{\dagger}(t) a(t) \rangle = \mathcal{E}^{2} \text{tr} \left[n \rho(t)\right]
$$(eq:intensity_quantum)

where $n=a^{\dagger} a$ is the number operator and $\rho(t)$ is the density operator of photons. Th expectation value $\langle \cdots \rangle \equiv \text{tr} (\cdots \rho)$ is an ensemble average.  At the last step we switched the picture from the the Heisenberg to Schr&ouml;dinger. 

Equation {eq}`eq:intensity_quantum` indicates that the intensity of light (in the mode we are looking at) is simply an average number of photons in the mode.    Be aware that we are counting only the photons in a particular mode, not necessarily all photons in the space.  _Full counting statistics depends on the selection of a mode._

## Quantum first-order coherence function

The classical first-order correlation function {eq}`eq:g1_classical`  can be easily extended to a quantum system in steady state.

$$
g^{(1)}(\tau) =  \frac{\langle E^{(-)}(t+\tau) E^{(+)}(t)\rangle_{ss}}{\langle E^{(-)}(t) E^{(+)}(t)\rangle_{ss}} = \frac{\langle a^{\dagger}(t+\tau) a(t)\rangle_{ss}}{\langle n \rangle_{ss}}
$$(eq:g1-qm-general)

where the expectation value is taken with the steady state.  While Eq. {eq}`eq:g1-qm-general` is expressed in the Heisenberg picture, actual calculation is done in the Schr&ouml;dinger picture with a trick based on quantum regression theorem, which we discussed in {numref}`sec:qme`.  It cab found easily that $g^{(1)}(0) = 1$.



## Quantum second-order coherence function

The extension of the classical the second-order coherence function to a quantum version is need a special care due to a rule of quantum field theory known as normal order, which guarantees that the vacuum state does not have energy in it.

$$
g^{(2)}(\tau) = \frac{\langle : E^{(-)}(t+\tau) E^{(+)}(t+\tau) E^{(-)}(t) E^{(+)}(t) : \rangle_{ss}}{\langle E^{(-)}(t+\tau) E^{(+)}(t+\tau)\rangle\langle E^{(-)}(t) E^{(+)}(t)\rangle_{ss}}
= \frac{\langle a^{\dagger}(t) a^{\dagger}(t+\tau) a(t+\tau) a(t)\rangle_{ss}}{\langle n \rangle_{ss}^2}
$$

where $: \cdots :$ indicate normal order.  Notice the specific order of operators.  The annihilation operators must go the the right and act on the ket directory due to normal ordering.  In addition, the operator must act in the chronological order according to the experimental procedure.

WE are particularly interested in the value at $\tau=0$.  Using the commutation relation, $a^{\dagger}a^{\dagger} a a = a^{\dagger} a a^{\dagger} a - a^{\dagger} a = n^2-n$, we find

$$
g^{(2)}(\tau) = \frac{\langle n(n-1) \rangle_{ss}}{\langle n \rangle_{ss}^2}
$$(eq:g2-qm-general)


## Practical theoretical investigation

While Eqs. {eq}`eq:g1-qm-general` and {eq}`eq:g2-qm-general` are useful in rigorous theoretical investigation,  we don't measure the number of photons in many practical cases.  In the following chapters, we use quantum master equations and quantum trajectories to compute the coherence functions. No photon number is actually computed or measured.  Instead, we measure the jumps of electrons in the light source and assume that at each jump a photon is emitted to a mode. If the detector is interacting with the same mode, it counts the photons in the mode.  Thus it is assumed that the emission counting is equivalent to photon counting.  We should keep in our mind that we are not actually calculating the photons detected by the detection devices.  If the optical mode the detector is measuring is different from the mode the light source is interacting, theoretical prediction can be misleading.  You will see such cases later in the case studies.

:::{warning}
There is an important remark in place.   Popular theoretical approaches seen in books and papers do not calculate photon counting statistics.  They actually calculate "emission statistics" by measuring the time intervals of electron deexcitation. When an electron is deexcited, a photon is emitted.  Hence, the emission counting statisrtics and photon counting statistics should be the same.  However, it is true only when all photons emitted by quantum dots are counted.  In reality, many experiments don't capture all photons.  They may be measuring photons emitted in a certain direction.  Then, there is difference between emission counting and actual photon counting.  In fact, experimental data show that both bunching and antibunching photons are simultaneously observed from a pair of atoms depending on the location of the measurement device.{cite:p}`Wolf2020`.
:::
