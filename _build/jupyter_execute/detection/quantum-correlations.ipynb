# Quantum correlation functions

In {numref}`sec:quantum-lights`, we learn three different types of photon counting statistics.  The key point was that the size of variance with respect to the mean determines the character of the light. We also see in {numref}`sec:classical-lights` that $g^(2)(0)$ dictates the variance of intensity at least for classical lights.  Hence, the correlation function $g^(2)(\tau)$ for quantum lights can be a good indicator of the photon counting statistics.  In this chapter, we derive a general mathematical expression of the correlation functions and discuss their general properties.  The case studies with simple models follows in separate sections.

We continue to assume that photons are in a single mode.

## Relation with photons and electric fields.

To derive the quantum version of {eq}`eq_g2-classical`, we define field operators corresponding to $\mathbf{E(t)}$ and $\mathbe{E}^{*}(t)$ in the Heisenberg picture as

$$
\begin{eqnarray}
E^{(+)}(t) &=& \mathcal{E}_0 a(t)= \mathcal{E}_0 a e^{-i \omega t} \\
E^{(-)}(t) &=& \mathcal{E}_0 a^\dagger(t) = \mathcal{E}_0 a^{\dagger} e^{i \omega t}
\end{eqnarray}
$$

where $a(t)$ and $a^\dagger(t)$ are annihilation and creation operators in the Heisenberg picture.  $\mathcal{E}_0$ is a proportional constant.  Since the photons are in a single mode, the direction of polarization is fixed and thus we consider only the field component in that direction.  Then, the intensity of the field is given by

$$
I(t) = \langle E^{(-)}(t) E^{(+)}(t)\rangle = \mathcal{E}_{0}^{2} \langle a^\dagger(t) a(t) \rangle = \text{tr} \left[n \rho(t)\right]
$$

where $n=a^\dagger a$ is the number operator and $\rho(t)$ is the density operator of photons.  At the last step we switched from the the Heisenberg picture to Schr&ouml;dinger picture. The final expression indicate that the intensity is just an expectation value of the photon number.


:::{note}
There is an important remark in place.   Popular theoretical approaches seen in books and papers do not calculate photon counting statistics.  They actually calculate "emission statistics" by measuring the time intervals of electron deexcitation. When an electron is deexcited, a photon is emitted.  Hence, the emission counting statisrtics and photon counting statistics should be the same.  However, it is true only when all photons emitted by quantum dots are counted.  In reality, many experiments don't capture all photons.  They may be measuring photons emitted in a certain direction.  Then, there is difference between emission counting and actual photon counting.  In fact, experimental data show that both bunching and antibunching photons are simultaneously observed from a pair of atoms depending on the location of the measurement device.{cite:p}`Wolf2020`.
:::
