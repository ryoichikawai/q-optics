# A driven TLS:  Rabi oscillation

In this chapter, we investigate a single two-level system (TLS) driven by an time-dependent external field. The system Hamiltonian is no longer diagonal in the energy basis due to the driving and thus the situation is slightly different from the example discussed in {numref}`sec:qme` and {numref}`sec:jump`. First we introduce a mathematical model.  Then, the time-evolution of the system and second order coherence are investigated by two different approaches, 1) open quantum system, and 2) quantum trajectory. 

The main difference from the simple example discussed in {numref}`sec:qme` is the presence of a driving term in the system Hamiltonian.  The main consequence is the appearance of the Rabi oscillation in the system evolution and in the first/second-order coherence.  Hence, the focus on this chapter is the Rabi oscillation.
