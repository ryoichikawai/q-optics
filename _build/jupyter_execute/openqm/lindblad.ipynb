# Quantum master equation (Lindblad equation)

The state of a system is described by a state vector $|\psi\rangle$ if it is a pure state or a density operator $\rho$ if it is a mixed state.  If the system is not fully isolated, the state is necessarily mixed and thus we must use the density operator.  In this section, we develop an approximated approach to find the time evolution of a density operator for a system interacting with a thermal reservoir.   Lindblad originally developed the method (the Lindblad equation) which is now commonly known as the quantum master equation. 

Using this methods, we can obtain for examples:
1. the evolution of the state $\rho(t)$ and the steady state as asymptotic limit $\rho^{ss} = \lim_{t \rightarrow \infty} \rho(t)$.
2. the expectation value of observable $A$ as $\langle A(t) \rangle = \text{tr}[A \rho(t)]$.
3. temporal correlation functions such as $g^{(1)}(\tau)$ and $g^{(2)}(\tau)$ with the help of quantum regression theorem.
4. quantum information using the von Neumann entropy $S(t) = \text{tr} [\rho(t) \ln \rho(t)]$.

and more.

However, it is not easy to obtain full counting statistics (in principle possible but very tedious.)  For that, there is a better method known as quantum trajectory simulation which we discuss in next chapter.

## Derivation of quantum master equation - a summary

Here a very brief summary of the derivation steps is provided so that we know what approximations are used.  Read a textbook listed in the previous page for the full derivation.
Consider a system S surrounded by a thermal reservoir R.  We assume that the reservoir is an ideal heat bath which remains in thermal equilibrium all time.  which is a key assumption to derive quantum master equation. The total system S+R is completely isolated and thus its time evolution is unitary.  The Hamiltonian of the total system can be divided to three parts:

$$
H_\text{S+R} = H_\text{S} \otimes I_\text{R} + I_\text{S} \otimes H_\text{R} + V_\text{SR}
$$

where $H_\text{S}$,  $H_\text{R}$ and $V_\text{SR}$ are the Hamiltonian for the system S, the reservoir R, and the interaction between S and R.  $I_\text{S}$ and $I_\text{R}$ are identity operators in each Hilbert space. 

The state of the total system is described by the density operator $\rho_\text{SR}$ and its time evolution is determined by the Liouville-von Neumann equation

$$
\frac{d}{dt} \rho_\text{SR} = -i \left[H_\text{SR}, \rho_\text{SR}\right].
$$

To solve this equation, two approximations are introduced.

1. The interaction between S and R is weak enough to use the Born approximation (1st order perturbation). Only $V_\text{SR}$ and (V_text{SR})^2$ are taken into account and all higher order terms are neglected.
2. The reservoir is always in the thermal state $\rho_\text{R} = \text{tr}_\text{S} \left( \rho_\text{SR} \right)= \frac{1}{Z_\text{R}} e^{-\beta H_\text{R}}$ where $\text{tr}_\text{S}$ is a partial trace over the system Hilbert space.  $\beta$ and $Z_\text{R}=\text{tr}_\text{R} e^{-\beta H_\text{R}}$ are usual inverse temperature and the partition function.
3. When the reservoir is disturbed, it _instantaneously_ relaxes to the original equilibrium state.  Hence, the state of reservoir remains in $\rho_\text{R}$ all time.  This implies that the reservoir has no memory of past events. This assumption is called the Markovian approximation. 
4. The coupling Hamiltonian is separable.  Mathematically, $V_\text{SR} = \sum_{i} A_i \otimes B_i$.  Further we assume that $\langle B_i \rangle = \text{tr} \left( B_{i} \rho_\text{R} \right) = 0$.
5. We introduce jump operators, each of which represents either excitation or deexcitation process.  

   $$
   A_{i} (\omega) = \sum_{k,k'} \delta(E_{k} - E_{k'} - \omega) |k\rangle \langle k | A_{i} |k' \rangle \langle k'|
   $$

   $A_{i}(\omega)$ and $A_{i}^{\dagger}(\omega)$ behave like annihilation and creation operators for $i$-th mode. From the completeness relation $\sum_{\omega} A_{i}(\omega) = \sum_{\omega} A_{i}^{\dagger}(\omega) = A_i$.
6. All the effects of the reservoir are taken into account through the temporal correlation of the reservoir operators.

   $$
   c_{ij}(\tau) = \text{tr}_\text{R} \left(B_j(\tau) B_i(0) \rho_\text{R}\right), \quad \tau>0.
   $$


7. Based on the Born-Markovian approximation, the state of the system can be approximated as $\rho_\text{SR} (t) \approx \rho_\text{S} (t) \otimes \rho_\text{R}$ where $\rho_\text{S} = \text{tr}_\text{R} \rho_\text{SR}$. 

At last we have a general form of quantum master equation.

$$
\frac{d}{dt} \rho_\text{S} = -i \left[ H_\text{S} + H_\text{LS}, \rho_\text{S}\right] + \mathcal{D}(\rho_\text{S})
$$(eq:qme_general)

where the $\mathcal{D}$ is known as dissipator defined by

$$
\mathcal{D}(\rho) = \sum_{\omega} \sum_{ij} \Gamma_{ij}(\omega) \left[ A_{j}(\omega) \rho A_{i}^{\dagger}(\omega) - \frac{1}{2} \left\{ A_{i}^{\dagger}(\omega)A_{j}(\omega), \rho \right\} \right]
$$(eq:dissipator_general)

where $\{\cdot,\cdot\}$ is anti-commutator and the dissipation rate $\Gamma_{ij}$ is the Fourier transform of the correlation function:

$$
\Gamma_{ij} (\omega) = \int_{-\infty}^{+\infty} c_{ij}(\tau) e^{i\omega \tau} d\tau.
$$

In Equation {eq}`eq:qme_general`, the unitary part include is a Lamb shift $H_\text{LS}$ defined by

$$
H_\text{SL} = \frac{1}{2i} \sum_{\omega} \sum{ij}  \left(\Gamma_{ij}(\omega) - \Gamma_{ji}^{*}(\omega)\right)
$$

In most applications, the Lam-Shift is small.  Hence, we shall omit it throughout this note.

The quantum master equation {eq}`eq:qme_general` works well only when the Born-Markovian approximation is valid.  In many applications in quantum optics, the approximation holds reasonably well.

To simplify mathematical notation, the master equation {eq}`eq:qme_general` is often expressed as

$$
\frac{d}{dt} \rho = \mathcal{L} \rho
$$

with  a super operator  $\mathcal{L}$ defined as

$$
\mathcal{L} = -i \left[ H_\text{S} + H_\text{LS}, \cdot \right] + \mathcal{D}(\cdot)
$$

and a symbolical solution to the equation is

$$
\rho(t) = e^{\mathcal{L} t} \rho(0).
$$

We shall utilize the simplistic expression from time to time.

By changing the basis, the $\Gamma_{ij}$ in Eq. {eq}`eq:dissipator_general` can be diagonalized.  The diagonal form of the disspator is written 

$$
\mathcal{D}(\rho) = \sum_{\omega} \sum_{i} \Gamma_{i}(\omega) \left[ L_{i}(\omega) \rho L_{i}^{\dagger}(\omega) - \frac{1}{2} \left\{ L_{i}^{\dagger}(\omega)L_{i}(\omega), \rho \right\} \right]
$$

which is the standard form shown in literature.  The operator $L_i$ is often called Lindblad operator or _jump operator_.  Yet another form is often used, in particular QuTiP.  The $\Gamma_i$ is absorbed into the definition of jump operators and the dissipator is written as  

$$
\mathcal{D}(\rho) = \sum_{\omega} \sum_{i} \left[ C_{i}(\omega) \rho C_{i}^{\dagger}(\omega) - \frac{1}{2} \left\{ C_{i}^{\dagger}(\omega)C_{i}(\omega), \rho \right\} \right]
$$

where $C_{i}(\omega) = \sqrt{\Gamma_i(\omega)} L_i(\omega)$, which we shall call _collapse operator_. 

## A simple example - optical master equation for a two-level atom

Consider a system with two energy levels, a ground state $| g\rangle$ of energy $E_{g}$ and  an excited state $|e\rangle$ of energy $E_{e}$. I shall call such a system a two-level atom in this note.  However, it can be a molecule, a quantum dot or some other systems in which only two energy levels are dominantly involved in the processes we are interested in and contribution from other levels is assumed to be negligibly small.

Lowering and raising operators are defined with $\ket{g}$ and $\ket{e}$ as basis  respectively by

$$
\sigma_{+} =  | e \rangle\langle g |  \qquad \sigma_{-} = | g \rangle\langle e |
$$(eq:sigma-plus-minus)

The standard Pauli operators are expressed in the same basis as

$$
\begin{eqnarray}
\sigma_{1} &=& \sigma_{+}+\sigma_{-} = | e \rangle\langle g |+| g \rangle\langle e | \\
\sigma_{2} &=& -i \left(\sigma_{+}-\sigma_{-}\right) = -i (| e \rangle\langle g | - | g \rangle\langle e |) \\
\sigma_{3} &=& | e \rangle\langle e | - | g \rangle\langle g |
\end{eqnarray}
$$

The Hamiltonian of the atom is given by

$$
H_\text{S} = \frac{\hbar}{2} \omega_{0} \sigma_{3}
$$

where $\hbar\omega_{0} = E_{e} - E_{g}$ is the excitation energy.

Since there is only one $\omega$, and two transitions, we have two modes, 

$$
\mathcal{D}_{a}(\rho) = \Gamma_{a} \left(\sigma_{+}\rho\sigma_{-} - \frac{1}{2} \sigma_{-}\sigma_{+}\rho - \frac{1}{2} \rho \sigma_{-}\sigma_{+} \right)
$$

for absorption and

$$
\mathcal{D}_{e}(\rho) = \Gamma_{e} \left(\sigma_{-}\rho\sigma_{+} - \frac{1}{2} \sigma_{+}\sigma_{-}\rho - \frac{1}{2} \rho \sigma_{+}\sigma_{-} \right)
$$ 
for emission.

The dissipation rates are

$$
\Gamma_{a} = \gamma_{0} N, \qquad \Gamma_{e} = \gamma_{0}(N+1)
$$

where  the rate of spontaneous emission is defined by

$$
\gamma_{0} = \frac{4 \omega_{0}^{3} |\vec{d}|^2}{3\hbar c^3}
$$

and the Planck distribution by

$$
N=\frac{1}{e^{\hbar \omega_0/T}-1}
$$

The jump operators are $A_{a} = \sigma_{-}$  for absorption and $A_{e} = \sigma_{+}$,  which are obvious from the definition of $\sigma_{\pm}$.

The final form of the quantum master equation  is 

$$
\frac{d}{dt} \rho = -i \frac{\omega_{0}}{2} [\sigma_{3},\rho] + \mathcal{D}_{a}(\rho) + \mathcal{D}_{e}(\rho)
$$(eq:qme_optical)

which is simply an ordinary differential equation of $2\times2$ matrix.  This is simple enough to solve analytically.

## Bloch equations

There are many ways to solve Eq. {eq}`eq:qme_optical`.  There is a special method that works only for the two-dimensional Hilbert space.  Recalling that the identity matrix $I$ and the Pauli matrices $\sigma_{i}$ form a basis set. Expanding the density operator in the basis set,

$$
\rho(t) = \frac{1}{2}\left(I + \vec{r}(t) \cdot \vec{\sigma} \right)
$$

where $\vec{r}(t) = (r_{1}(t),r_{2}(t),r_{3}(t))$ is the Bloch vector, which uniquely determine the density operator.{cite:p}`Breuer2002,Lambropoulos2007`  Since the density operator is self-adjoint, $\vec{v}$ is real.  In fact, it is easy to show that $\vec{r} = \langle \vec{\sigma}\rangle = \text{tr}\left[\vec{\sigma}\rho(t)\right]$.  Multiply $\vec{\sigma}$ to Eq. {eq}`eq:qme_optical` and taking trace, we obtain the equation of motion for the Bloch vector is given by

$$
\begin{eqnarray}
\frac{d}{dt} \langle \sigma_1(t)\rangle &=& \omega_0 \langle \sigma_2 (t)\rangle - \frac{\gamma}{2} \langle \sigma_1(t)\rangle \\
\frac{d}{dt} \langle \sigma_2(t)\rangle &=& \omega_0 \langle \sigma_1 (t)\rangle - \frac{\gamma}{2} \langle \sigma_2(t)\rangle \\
\frac{d}{dt} \langle \sigma_3(t)\rangle &=& - \gamma \langle \sigma_1(t)\rangle  - \gamma_{0}
\end{eqnarray}
$$(eq:Bloch_eq)

where $\gamma = \gamma_0 (2N+1)$.  Equation {eq}`eq:Bloch_eq` is called the Bloch equation.{cite:p}`Breuer2002,Meystre2007`

By letting the left hand side 0, we can immediately find the steady state.

$$
\langle \sigma_1\rangle_{ss} = \langle \sigma_2\rangle_{ss} = 0, \qquad \langle \sigma_3\rangle_{ss} = -\frac{\gamma_0}{\gamma} = -\frac{1}{2N+1}
$$

The probability to find the atom in the excited state is given by

$$
p_{e}^{ss} = \frac{1}{2} \left(1 + \langle \sigma_3\rangle_{ss}\right) = \frac{N}{2N+1} = \frac{1}{Z_R}e^{-\beta \omega_0}
$$(eq:p_e_ss)

which is just the thermal equilibrium state.  Hence, the system is relaxed to the thermal state.  


Next, we calculate the relaxation to the thermal state starting from a pure state $|\psi_0\rangle = \cos \theta |g\rangle + \sin \theta |e\rangle$, which gives the initial density operator

$$
\rho(0) = |\psi_0\rangle\langle \psi_0| = \cos^2\theta |g\rangle\langle g| + \cos\theta\sin\theta |g\rangle\langle e| + \cos\theta\sin\theta |e\rangle\langle g| + + \sin^2\theta |e\rangle\langle e| 
$$

The initial condition for the Bloch equation is 

$$
\langle \sigma_1(0) \rangle =\sin (2\theta), \quad \langle \sigma_2(0) \rangle = 0, \quad \langle \sigma_3(0) \rangle = \cos(2\theta)
$$

The solution is 

$$
\begin{eqnarray}
\langle \sigma_1(t) \rangle &=& e^{\gamma t/2} \cos(\omega_{0} t) \sin(2\theta) \\
\langle \sigma_2(t) \rangle &=& e^{\gamma t/2} \sin(\omega_{0} t) \sin(2\theta) \\
\langle \sigma_3(t) \rangle &=& e^{\gamma t}  \cos(2\theta)  + (1- e^{-\gamma t}) \rangle \sigma_3(t) \langle_{ss}
\end{eqnarray}
$$

which indicates $\gamma$ is the relaxation rate.

A nice things about the Bloch equation is that you can visualize the solution on the Bloch sphere.  We will see it in the numerical example.

## Solving a master equation numerically using QuTiP

We solve the master equation 

```python
# ignore warnings (qutip issues various unwanted warnings)
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import numpy as np
from qutip import *

# Parameters
omega0 = 1.0   # excitation energy
gamma0 = 0.1   # spontaneous emission rate
temperature = 1.0  # tempeature of photon gas
N = 1/(np.exp(omega0/temperature)-1) # Planck distribution
theta = 2*np.pi/10 # initial condition

gamma1 = gamma0 * N   # coefficient to absorption
gamma2 = gamma0 * (N+1)  # coefficient to emission

# Hamiltonian for a two-level system (in the rotating frame)
H = 0.5 * omega0 * sigmaz() 

# Jump operators
c_ops = [np.sqrt(gamma1) * sigmap(),np.sqrt(gamma2) * sigmam()]

# Initial state: start in the ground state |g>
psi0 = basis(2,0)*np.sin(theta) + basis(2,1)*np.cos(theta)


# simulation time
times = np.linspace(0, 40, 400)

# observables to compute
observables = [sigmax(),sigmay(),sigmaz()]

# solve the Lindblad Equation
result = mesolve(H, psi0, times, c_ops, observables)

# probability to find the atom in the excited state
# p = (1+result.expect[2])/2

# 4. Plotting the Time Evolution
plt.figure(figsize=(4, 3))
plt.plot(times,result.expect[0], label=r"$\sigma_x$")
plt.plot(times,result.expect[1], label=r"$\sigma_y$")
plt.plot(times,result.expect[2], label=r"$\sigma_z$")
plt.title("Solution to the Lindblad equation")
plt.xlabel("Time")
plt.ylabel("<sigma_z>")
plt.legend(loc=1)
plt.grid(True)
plt.show()
```

We plot the results in the Bloch sphere.  The state is initially on the surface of the Bloch sphere since it is a pure state.  Then, the state spirals down to the thermal equilibrium located on the $z$ axis.

```python
plt.figure()
b = Bloch(figsize=(3,3))
b.add_points(result.expect, meth='l')  # 'l' adds a line between points
b.show()
plt.close()
```

## Calculation of $g^{(2)}(\tau)$

An extra step i sneeded to compute $g^{(2)}(\tau)$.  By definition, $g^{(2)}(\tau)$ measures the correlation of photon intensity.  But we are not calculating photons.  How can we calculate it?  When the transition from $|e\rangle$ to $|g\rangle$ takes place, a photon is emitted.  Therefore, if we know the statistics of the  emission transition, we have statistics of photons assuming that we measure all photons coming out from the atom. In other words, the optical mode we measure is all photons emitted by the atom.

Since the creation of photon is equivalent to the deexcitation of the atom the filed opetators can be expressed as follows:

$$
E^{(+)}(t) = \mathcal{E}_0 a(t) \propto  \sigma_{-}, \qquad E^{(-)}(t) = \mathcal{E}_0 a(t) \propto  \sigma_{+},
$$

and thus th eoperator corresponding to the intensity is

$$
I (t) = E^{(-)}(t) E^{(+)}(t) = \mathcal{E}_0^2 a^{\dagger}(t) a(t) \propto \sigma_{+}(t) \sigma_{-}(t)
$$

The correlation functions are now defined as

$$
g^{(1)}(\tau) = \frac{\langle \sigma_{+}(\tau) \sigma_{-}(0)\rangle_{ss}}{\langle  \sigma_{+}(0)\sigma_{-}(0) \rangle_{ss}}
$$(eq:g1-two-level-atom)

and

$$
g^{(2)}(\tau) = \frac{\langle :I(\tau)I(0): \rangle_{ss}}{(\langle I(0) \rangle_{ss})^2} = \frac{\langle \sigma_{+}(0)\sigma_{+}(\tau)\sigma_{-}(\tau) \sigma_{-}(0) \rangle_{ss}}{(\langle  \sigma_{+}(0)\sigma_{-}(0) \rangle_{ss})^2}
$$(eq:g2_two-level-atom)

where $:\cdots:$ indicates normal order of creation and annihilation operators and $\langle \cdots\rangle_{ss}$ average over the steady state.

There is a problem with the definitions {eq}`eq:g1-two-level-atom` and {eq}`eq:g2_two-level-atom`.  We don't how to calculate the expectation value  $\langle \cdots\rangle_{ss}$ since two different times are involved. There is a trick that works only when the Born-Markovian approximation is valid as discussed next.

:::{note}
There is an important remark in place.   Popular theoretical approaches seen in books and papers do not calculate photon counting statistics.  They actually calculate "emission statistics" by measuring the time intervals of electron deexcitation. When an electron is deexcited, a photon is emitted.  Hence, the emission counting statisrtics and photon counting statistics should be the same.  However, it is true only when all photons emitted by quantum dots are counted.  In reality, many experiments don't capture all photons.  They may be measuring photons emitted in a certain direction.  Then, there is difference between emission counting and actual photon counting.  In fact, both bunching and antibunching photons are simultaneously observed from a pair of atoms depending on the location of the measurement device due to quantum interference effects.{cite:p}`Wolf2020`.  If all photons from the pair were counted, bunching and anti-bunching would be cancelled out.
:::

## Quantum regression theorem


Suppose that the expectation value of an operator $A$ evolves in time. It should be written as $\langle A \rangle_t$ (expectation evaluated at time $t$) but  often written as $\langle A(t) \rangle$ which suggests that $A$ depends on time. If the system evolution is unitary, then $A(t)$ can be considered as an operator in the Heisenberg picture and the average is taken with respect to an initial state.  However, when the evolution of the system is not unitary, we can't define the usual Heisenberg picture. We interpret it as $\langle A(t) \rangle \equiv \text{tr}(A \rho(t))$  where the state is evolving in time and the operator is stationary.

A problem arises when operators depend on two different times.  For example, consider a correlation function $\langle A(t)B(0) \rangle$. What does $\langle \cdots \rangle$ mean in this case?   The quantum regression theorem states that

$$
\langle A(\tau),B(0) \rangle = \text{tr} [ A e^{\mathcal{L}t} (B \rho(0)) ]
$$

Notice that neither $A$ nor $B$ depends on time. $B \rho(0)$ is evaluated at $t=0$ and evolves to t=$\tau$ by the master equation.  Then we take the trace.It is a kind of surprising that the same evolution operator $e{\mathcal{L}t}$ can be used to evolve $B\rho$.  We can calculate $e^{\mathcal{L} t} (B \rho(0))$ by solving the differential equation

$$
\frac{d}{dt} \eta = \mathcal{L} \eta
$$

with the initial condition  $\eta(0) = B\rho(0)$.

Using the quantum regression theorem, the first order correlation can be computed as

$$
g^{(1)}(\tau) = \text{tr} [ \sigma_{+} e^{\mathcal{L}t} (\sigma_{-} \rho_{ss}] / \text{tr} [\sigma_{+}\sigma_{-} \rho_{ss}]
$$

and for the second order we have

$$
g^{(2)}(\tau) = \text{tr} [ \sigma_{+} \sigma_{-} e^{\mathcal{L} \tau} ( \sigma_{-} \rho_{ss} \sigma_{+}) ] / \left(\text{tr} [\sigma_{+}\sigma_{-} \rho_{ss}]\right)^2
$$
where we have rotated the operators inside the trace.  

Substituting Eqs. {eq}`eq:sigma-plus-minus`, we find an explicit expression of the correlation function

$$
g^{(2)}(\tau)= \frac{\langle e | \left(e^{\mathcal{L}\tau} |g\rangle\langle g |\right) | e \rangle}{\langle e | \rho_{ss} | e \rangle}  = \frac{\langle e | \rho(\tau;g) | e \rangle}{\langle e | \rho_{ss} | e \rangle} = \frac{p_{e}(\tau;g)}{p_{e}^{ss}}
$$(eq:g2_explicit)

where $\rho(\tau;g)$ is a state at $t=\tau$ starting from the ground state at $t=0$ and $p_{e}(\tau;g)$ is a conditional probability that the system is in the ground state at $t=0$ and in the excited state at $t=\tau$.  $p_{s}^{ss}$ is the probability of the steady state being in $|e\rangle$ whci we already know from {eq}`eq:p_e_ss`.  Noting that $\lim_{\tau \rightarrow \infty} p_{e}(\tau;g) = p_{e}^{ss}$, we find $\lim_{\tau \rightarrow \infty} g^{(2)}(\tau) = 1$.  On the other hand, $\lim_{\tau \rightarrow 0} p_{e}(\tau;g) = 0$ since $|e\rangle$ and $|g\rangle$ are orthogonal.  Thus we have $g^{(2)}(0) = 0$.

From the final expression of Eq. {eq}`eq:g2_explicit`, only we need is $p_{e}(\tau;g)$ which can be computed by solving the master equation.




The QuTiP package contains a module that do every thing for us. We just provide the information of the quantum master equation, namely the Hamiltonian and jump operators.

```python
# Continued from the previous code block. 
# You need to execute it before running this code.

# We already set up Hamiltonian H, collapse operators c_ops
# in the first code block in this page.
# We need to modify
# 1) psi0 - initial state
# 2) operator to be measured. that is sigmaz

psi0 = basis(2,1) 
result = mesolve(H, psi0, times, c_ops, sigmaz())

pe_tau = (1+result.expect[0])/2
pe_ss = N/(2*N+1)

g2 = pe_tau/pe_ss

plt.figure(figsize=(4,3))
plt.plot(times, g2,c='k')
plt.plot(-times,g2, c='k')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$g^{(2)}(\tau)$')
plt.grid(True)
plt.show()
```
