# Open quantum system approaches

The standard quantum mechanics was developed for closed (isolated) systems, hence the total energy must be conserved.  In a real world, a system is surrounded by environments.  It is difficult to completely isolate the system. The interaction between the system and the environments plays significant roles even when the interaction is weak.   On the other hand, it is difficult to explicitly take into account the large degrees of freedom in the the environments. We need a theory which take into account the effect of the environment without solving its dynamics.  Thermodynamics is such a theory in which the effects of the environments (heat baths) is taken into account through heat and entropy (the 1st and 2nd laws of thermodynamics).  

In a closed system, a system evolve unitarily according to the Shcr&ouml;dinger equation $i\frac{d}{dt}|\psi\rangle = H |psi\rangle$ or equivalently  the Liouville-von Nuemann equation $ \frac{d}{t} \rho = -i [H, \rho]$.  We use the latter since the state of open quantum system is mixed in general. If we consider the system (S) and environment (E), the total system obey the unitary evolution:

$$
\frac{d}{dt} \rho_\text{S+E} = -i [H_{S+E},\rho_\text{S+E}]
$$

But we want to know only the system density $\rho_\text{S} = \text{tr}_\text{E} \left(\rho_\text{S+E}\right)$ where $\text{tr}_{E}$ is partial trace over the environmental degrees. $\rho_\text{S}$ is known as reduced density.   The theory of open quantum system tries to find $\rho_{S}$ without solving the evolution of the full system $\rho_\text{S+E}$ using certain approximations.  Here we briefly review the two most popular approaches, the Lindblad equation and stochastic jump method (quantum trajectory method).
