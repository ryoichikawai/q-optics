# Quantum correlation functions

In {numref}`sec:quantum-lights`, we learn three different types of photon counting statistics.  The key point was that the size of variance with respect to the mean determines the character of the light. We also see in {numref}`sec:classical-lights` that $g^(2)(0)$ dictates the variance of intensity at least for classical lights.  Hence, the correlation function $g^(2)(\tau)$ for quantum lights can be a good indicator of the photon counting statistics.  In this chapter, we derive a general mathematical expression of the correlation functions and discuss their general properties.  The case studies with simple models follows in separate sections.



## What is a mode?

What is a mode of photons?  It is a state of photons?  In many literature, it is not clear how a mode is defined.  To understand the concept of modes, we first look at the standard quantum mechanics.  A state of quantum particles is mathematically described by a vector $|\psi\rangle$ in a Hilbert space.  To specify a particular state, we use a basis set $\{|u_i\rangle, i=1, \cdots, N\}$ (a finite dimention $N$).  The state is now expressed as a superposition of the basis vectors as $|\psi\rangle = \sum_i c_i |u_i\rangle$ where $c_i$ is an expansion coefficient.  The basis set is not associated with the state and we can pick any basis set to express the state.  

Similarly, in quantum optics, a mode picks a particular "basis" in a Fock space by defining creation and annihilation operators $a$ and $a^{\dagger}$. When we need more than one mode, we have a set of creation and annihilation operators $\{a_k, a_k^{\dagger},k=1,\cdots,M\}$ where $M$ is the number of modes.  Like a basis set expansion in the Hilbert space, a state of photons are created by applying creation and annihilation operators on the vacuum state, like $|\psi\rangle=f(a,a^{\dagger})|0\rangle$ with a single mode. Then we say that the photons are in a single mode.  For example,  a so-called squeezed state of a single mode can be constructed 

$$
|\psi\rangle = e^{\frac{1}{2}\left(\xi^{*} a^{2} - \xi (a^{\dagger})^2\right)} |0\rangle.
$$

where $\xi$ and $\xi^{*}$ are parameters defining a particular state similar to the expansion coefficients $c_i$.

Such construction of a state does not make a sense until $a$ and $a^{\dagger}$ are defined. Now, a question is how to define them.  The choice is up to us.  We pick a particular $a$ and $\{dagger}$ suitable for our objective.  In a certain experimental setup, one that creates a wave packet in space with a certain frequency spectrum may be convenient. For photons in an optical cavity, the choice seems obvious.  For theoretical investigation, one that diagonalizes a density operator may be ideal.   

Utilizing the freedom to choose modes, we focus on photons in a single mode relevant to experimental setups we are interested in. 


## Relation with photons and electric fields.

To derive the quantum version of {eq}`eq_g2-classical`, we define field operators corresponding to electric field $\mathbf{E(t)}$ and $\mathbf{E}^{*}(t)$ in a single mode.  We need to define the mode first.  We choose modes $\{a_{\omega}\}$ such that the free Hamiltonian of photons is given by $H_{0} = \sum_{\omega} \frac{\omega}{2}a^{\dagger}_{\omega} a_{\omega}$.  The experimental setup selects one mode with a particular frequency by some filtering mechanism.  If the frequency spectrum is continuous, there is a mathematical issue in selecting one particular frequency.  Then, we define a mode with frequency $\omega_0$ as

$$
a_{f} = \int d\omega f(\omega) a_{\omega}
$$

where $f(\omega)$ is a band filter with normalization $\int d\omega |f(\omega)|^2=1$ and mean frequency $\int d\omega~ \omega |f(\omega)|^2=\omega_0$.  The normalization guarantees the boson commutation relation $[a_{f},a_{f}^{\dagger}]=1$.   We assume that the band width is sufficiently narrow that $a_{f}$ is a well defined mode with a frequency $\omega_0$ and the Hamiltonian is effectively $H_0 = \frac{\omega_0}{2} a_{f}^{\dagger}a_{f}$. The operators $a_{f}$ and $a_{f}^{\dagger}$ destroys and creates photons in the mode, respectively.   For multi modes, there is a formal way to construct band filtering functions  but we don't discuss it here.

Now, we express the electric field corresponding to the filtered mode $a$ (to to simplify expressions $a_f \rightarrow a$ and $\omega_0 \rightarrow \omega$ hereafter) by

$$
\begin{eqnarray}
E^{(+)}(t) &=& \mathcal{E} a(t)= \mathcal{E}_0 a e^{-i \omega t} \\
E^{(-)}(t) &=& \mathcal{E} a^{\dagger}(t) = \mathcal{E}_0 a^{\dagger} e^{i \omega t}
\end{eqnarray}
$$

where $a(t)$ and $a^{\dagger}(t)$ are annihilation and creation operators in the Heisenberg picture and $a$ and $a^{\dagger}$ are corresponding operators in the Schr&ouml;dinger picture.     $\mathcal{E}$ is a proportional constant.  Since the photons are in a single mode, the direction of polarization is fixed and thus we consider only the field component in that direction.  Then, the intensity of the field is given by

$$
I(t) = \langle E^{(-)}(t) E^{(+)}(t)\rangle = \mathcal{E}^{2} \langle a^{\dagger}(t) a(t) \rangle = \mathcal{E}^{2} \text{tr} \left[n \rho(t)\right]
$$(eq:intensity_quantum)

where $n=a^{\dagger} a$ is the number operator and $\rho(t)$ is the density operator of photons. Th expectation value $\langle \cdots \rangle \equiv \text{tr} (\cdots \rho)$ is an ensemble average.  At the last step we switched from the the Heisenberg picture to Schr&ouml;dinger picture in which the density operator evolves in time. 

Equation {eq}`eq:inrtensity_quantum` indicates that the intensity of light (in the mode we are looking at) is simply an average number of photons in the mode.    Be aware that we are counting only the photons in a particular mode, not necessarily all photons in the light beam.  _Full counting statistics depends on the selection of a mode._
