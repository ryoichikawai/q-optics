# Quantum lights



Classical light is described by electromagnetic fields which are continuous both in space and time.  Quantum light is described also by electromagnetic fields but they are quantum fields.  The quantum fields are also continuous.  So, where is "quantumness" in the quantum light?  Photons immediately pops up in our head.  But what are they?  Unfortunately, they are poorly defined in literature  and misused by many people. In 1951 Albert Einstein said "All these fifty years of conscious brooding have brought me no nearer to the answer to the question 'What are light quanta?' Nowadays every Tom, Dick and Harry thinks he knows it, but he is mistaken."{cite:p}`French1979`  Since a correct use of the photon concept is so difficult, W. E. Lamb suggested that a license be required for use of the word photon."{cite:p}`Lamb1995`  Roy J. Clauber, who were given a license, defined photon as  “A photon is what a photodetector detects” and also "A photon is where the photodetector detects it" rather than defining a photon as a pre-existing "particle" in space.  When three Nobel laureates in quantum optics were saying like that, the meaning of photons must be very difficult for laymen to understand.

Although we don't have a license, we shall use the word "photon" in this note within a limited scope as discussed below.


## Optical modes

To define photons, first we need to define optical modes. But what are they?  Mathematically, they are a set of basis vectors in a Hilbert space.  Recall that a classical electromagnetic field, which is a solution to the Maxwell equation, is a vector in a vector space of infinite dimension.  In other words, the filed is expressed as a linear combination of basis functions $\{u_{i}(x,t)\}$:

$$
\mathbf{E}(x,t) = \sum_{j} \mathbf{e}_{j} (a_{j} u_{j}(x,t) + a^{*}_{j} u^{*}_{j}(x,t))
$$(eq:mode_decomposition)

where $a_{j}$ and $\mathbf{e}_{j}$ are an expansion coefficient and a polarization vector associate with the $j$-th mode. In a free space, $u_{j}(x,t) = e^{i(\mathbf{k}_{j} - \omega_{j})}$ is commonly used. We have used summation rather integration to avoid mathematical complication but in the real world, we have to use a continuous spectrum.

In principle, the choice of modes are arbitrary. We can pick modes that are convenient for the physical situation we are investigating. A problem is that a convenient choice of modes to describe a light source is not necessarily a good choice for the setup of photon detection.[^relativistic_mode].  It is very important to pick modes consistent to the experimental setup. We shall discuss it in chapters of case studies.

[^relativistic_mode]: The situation is more complicated in a relativistic case where the source and observer are not in the same inertial frame but that is not our concern in this note.

## Quantization of electromagnetic fields

Once we have a set of modes, field quatization is a straight forward. The coefficients $a_{j}$ and $a^{*}_{j}$ become annihilation operator $\hat{a}_{j}$ and creation operators $\hat{a}^{\dagger}_{j}$, respectively. They satisfy commutation relation $[\hat{a}_{i},\hat{a}^{\dagger}_{j}] = \delta_{ij}$ and follow the usual operator algebra of a Harmonic oscillator.  The Hamiltonian is given by $\sum_{j} \frac{\omega_{j}}{2} \left(a^{\dagger}_{j} a_{j} + \frac{1}{2}\right)$.
We define photons using the creation and annihilation operators. Defining a vacuum state by $a_{j}|0\rangle=0$, we create a photon in  $j$-th mode by $|1_j\rangle = a^{\dagger}_{j}|0\rangle$.  $N_{j} = a^\dagger_{j}a_{j}$ is a number operator as usual.  It is important to know that the operators are associated with the specific modes  we picked.  When different modes are used, we have different operators.   Foe example, consider a Bogoliubov transformation of basis from $u_{j}$ to $v_{k}$:

$$
v_{k} = \sum_{j} \left( \alpha_{kj} u_{j} + \beta_{kj} u_{j}^{*}\right).
$$

The annihilation operator for the new mode $v_{k}$ is given by

$$
\hat{b}_{k} = \sum_{j} \left( \alpha_{kj}^* \hat{a}_{j} - \beta_{kj}^* \hat{a}_{j}^{\dagger} \right).
$$

Note that the annihilation of photon in mode $v_{k}$ involves the creation of photon in mode $u_{j}$.  In general number of photons is not invariant under the transformation. This is quite confusing.  Are photons really  particles?

It is tempting to think that photons are particles like electrons.  They are NOT.  There is no position operator for photons, indicating that they are not point particles. Since the operators are tied to the modes, the number of photons also depends on the choice of modes.  The bottom line is that the quantum electromagnetic fields are fundamental but photons are emergent.  Some people call photon a pseudo particle.  This is why Glauber said "a photon is what a photodetector detects", which implies that we should not think about photons traveling around in the space. 

Until a license is given, we shall not discuss the details of photons in this note.  
We will mostly deal with a well-defined single mode and assume that both the light source and the detector are interacting with the same mode.  We use the word in two different occasions.  When a photodector records a "click", we say a photon is detected.  When an electron in a light source jumps from a higher energy state to a lower energy state, we say a photon is emitted.  What these statements mean is that the electron deposits a quanta of energy to the field, and then the energy stored in the field somehow travels to the detector and gives a click. That is all we need for now.
