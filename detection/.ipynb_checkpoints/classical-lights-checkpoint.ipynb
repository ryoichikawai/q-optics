# Detection of classical lights

## Intensity of classical lights

In the classical theory, light is electromagnetic wave which is continuous in space and time. Mathematically, it is expressed as vector field $\mathbf{E}(\mathbf{x},t)$ and $\mathbf{B}(\mathbf{x},t)$ Its character is measured with two temporal correlation functions, the first-order and second-order correlation defined respectively by

$$
g^{(1)}(\tau) = \frac{\langle\mathbf{E}^{*}(t) \cdot \mathbf{E}(t+\tau)\rangle_t}{\langle\mathbf{E}^{*}(t) \cdot \mathbf{E}(t)\rangle_t} = \frac{\langle\mathbf{E}^{*}(t) \cdot \mathbf{E}(t+\tau)\rangle_t}{\langle I(t) \rangle_t}
$$(eq_g1-classical)

and

$$
g^{(2)}(\tau) = \frac{\langle\left[\mathbf{E}^{*}(t) \cdot \mathbf{E}(t)\right]\left[\mathbf{E}^{*}(t+\tau) \cdot \mathbf{E}(t+\tau)\right]\rangle_t}{(\langle\mathbf{E}^{*}(t) \cdot \mathbf{E}(t)\rangle_t)^2}
= \frac{\langle I(t) I(t+\tau)\rangle_t}{(\langle I(t) \rangle_t)^2}
$$(eq_g2-classical)

where $I(t)=\mathbf{E}^{*}(t)\cdot \mathbf{E}(t)$ is the intensity of the field and $\langle\cdots\rangle_t$ indicates average over time $t$.   We have use the complex expression of the electric fields for mathematical convenience.

Usually, the correlation vanishes for a large $\tau$ and $\langle\mathbf{E}^{*}(t) \cdot \mathbf{E}(t+\tau)\rangle_t \rightarrow (\langle \mathbf{E}(t) \rangle_t)^2 = 0$ and $\langle I(t) I(t+\tau)\rangle_t = (\langle I(t) \rangle_t)^2$, from which we obatin.  

$$
\lim_{\tau\rightarrow\infty} g^{(1)}(\tau) = 0
$$

and

$$
\lim_{\tau\rightarrow\infty} g^{(2)}(\tau) = 1
$$

:::{note}
It is very important to keep it in our mind that the measurement is done at a certain fixed point in space, which is not explicitly written in the above expression. $g^{(1)}(\tau)$ and $g^{(2)}(\tau)$ can 
strongly depend on the location.
:::

The fist-order correlation provides information about the oscillatory behavior of the field through its Fourier transform known as power spectrum (Wiener–Khinchin theorem):

$$
I(\omega) = \frac{1}{2\pi} \int_{-\infty}^\infty g^{(1)}(\tau) e^{-i \omega \tau} d\tau
$$

In other words, the first-order correlation gives us frequency distribution of the light.

## No-delay correlation

No-delay ($\tau=0$) second-order correlation $g^(2)(0)$ becomes a very important quantity to investigate both experimentally and theoretically. 
While $g^{(1)}(0)=1$ is always true, the value of $g^{(2)}(0)$ varies depending on the character of light.  Introducing mean intensity $I = \langle I(t) \rangle_t$ and the deviation from it as 

$$
(\Delta I)^{2} = \langle I^{2}(t) \rangle_t - I^{2}
$$

we obtain

$$
g^{(2)}(0) = \frac{\langle I^{2}(t)\rangle_t}{(\langle I(t) \rangle_t)^2} = 1 + \left(\frac{\Delta I}{I}\right)^{2}
$$

which suggests that $g^{(2)}(0) \ge 1$. 

   

The positive correlation suggests that  This was first confirmed by Hanbury Brown and Twiss.{cite:p}`HanburyBrown1956`  with the advent of modern experimental technique,  cleaner results have been obtained.{cite:p}`Tan2014,Tan2017`    If and only if $\Delta I =0$, then $g^{(2)}(0)=1$.  This means that if an experiment data shows $g^{(2)}(0)=1$ then the light intensity is not fluctuating.
It turns out that negative value of $g^{(2)}(0)$ has been observed{cite:p}`Kimble1977,Grangier1986` in violation of the above result.  Such results can be explained only by quantum light.


## A little history of quantum optics

In 1954, Hanbury Brown and Twiss proposed a new type of interferometer.{cite:p}`HanburyBrown1954` Light from a single source is split in half, a half light goes to a detector and the other half to another detector.  The time-delayed correlation of the intensity fluctuation is measured between the two detectors. In 1956, they carried out the proposed experiment and observed strong correlation between the two detectors, especially at no time-delay.{cite:p}`HanburyBrown1956`  The result is known as Hanbury Brown and Twiss effect.  Essentially, this experiment confirmed that $g^(2)(0) > 1$. The results were seemingly obvious from the classical theory since the same light just took two separate paths.  But in 1950s, physicists were all excited by the success of quantum mechanics including photons. Many of them learned quantum mechanics from  Dirac's book "The principle of Quantum Mechanics" which said  "Each photon interferes only with itself."  Based on this, the results of Hanbury Brown and Twiss experiment was unacceptable to them. The two detectors are measuring two different group of photons which are independent to each other. 

The controversy was resolved by Fano{cite:p}`Fano1961`.  Two photons can be coherent if they are indistinguishable.  Suppose that photons $a$ and $b$ are detected respectively by detectors $A$ and $B$.  The probability that happens can be written as $|\langle AB| ab\rangle|^2$. There is another possiblity that photon $a$ goes to $B$ and $b$ to $A$. In other words, the state of photon must be \frac{1}{\sqrt{2}} \left(|ab\rangle + |ba\rangle$ since they are bosons.  Then, the probability that the two detectors find photons are

$$
\frac{1}{2} \left | \langle AB | (|ab\rangle + |ba\rangle ) \right |^2 = \frac{1}{2} \left(|\langle AB|ab\rangle|^2 + |\langle AB|ba\rangle|^2 \right) + \text{Re} \langle AB | ab \rangle \langle ba | AB\rangle
$$

Due to the interference term, the probability increases.   Therefore, two photons interfere themselves. Notice that the Dirac's statement is not violated. Note also that if the particles were fermions, then the inteference term is negative and thus the correlation is reduced, which has been confirmed.

Immediately after Hanbury Brown and Twiss physicists realized that a better theory on the coherence of light was urgently needed and modern quantum optics was born.{cite:p}`Glauber1969`. While the Hanbury Brown and Twiss effect can be easily explained by the classical theory of light, it initiated the modern quantum optics and even now their method is used.
