# Quantum light


Classical lights is described by electromagnetic fields which continuous both in space and time.  Quantum lights is described also by electromagnetic fields but they are quantum fields.  The quantum fields are also continuous.  So, where is "quantumness" in the quantum light?  Mathematically speaking, the difference between classical and quantum electromagnetic fields is that the latter is quantum mechanical operators, which is subject to quantum mechanical measurement. To construct the quantum fields concepts of "modes" and "photons" are invoked.  However, they are poorly defined in many books and misused by many people. Albert Einstein said "These days, every Tom, Dick and Harry thinks he knows what photon is, but he is wrong."   W. E. Lamb suggested that a correct use of the photon concept is so difficult that a license be required for use of the word photon."{cite:p}`Lamb1995,oychoudhuri2017`  Only a few, including Glauber, were given a licence.   When two Nobel laureates in quantum optics were complaining like that, the meaning of photons must be very difficult for laymen to understand.

## What is a photon?

The concept of photon is tightly connected to the concept of modes.  The basic approach in quantum optics is to set up "modes" and define creation and annihilation operators for each mode.  We might want to say that the creation operator creates a photon.  The problem of this picture is that the choice of the modes is not unique.  Changing the choice of modes splits the photons, suggesting that there is no photon at the ontological level. Some people call it pseudo-particle.   For example, if a light source generates "photons" in a certain mode but a detector is set to measure "photons" in a different mode, what happens. The detector records "click". Is that the same photon the source emitted?    In a free space, a choice of modes is completely arbitrary.  On the hand hand, there is no choice for single mode optical cavity.

In this lecture note, I will not enter the fundamental issue of photons. If you are interested in the issue, read"{cite:p}`Lamb1995,oychoudhuri2017` In this note we follow what Glauber said:  "A photon is what a photodetector detects."  The "click" is a photon and before that we have only light (electric fields).  A light is a smooth continuous wave but when a detector tries to measure it, a train of clicks are recorded. I shall call it a spike train of photons.  

In practice, we don't calculate the spike train.  We often rather calculate the state of electrons in the light source. When an electron jumps from a higher energy state to a lower energy state, a quanta of energy is leased to the electromagnetic field.
