# Photon Counting Statistics

## Three types of statistics

When a photon arrives at a detector, its arrival time is recorded.  After many photons arrived, the history of photon arrival is recorded as a time sequence. We are interested in the time intervals between two successive photon arrivals.  The statistical distribution of the intervals tells us how the photons are generated.

1. Poisson distribution:   Photons arrive independently at a constant average rate, and the variance in counts equals the mean.
2. Super-poissonian distribution:  Photons tend to bunch together, leading to a variance that is larger than the mean.
3. Sub-poissonian distribution: Photons arrive more regularly than random (antibunching), resulting in a variance smaller than the mean.
